###find maximum infection rate for India & China

```python
df_new.loc["India"].diff().max()
```

```python
df_new.loc["China"].diff().max()
```

###find maximum infection rate for all of the countries

```python
df_new.index
```

```python
countries = list(df_new.index)

max_infection_rates = []

for i in countries :

    max_infection_rates.append(df_new.loc[i].diff().max())

max_infection_rates
```

```python
df_new["max_infection_rates"] = max_infection_rates
df_new.head(10)
```

###Create a new dataframe with only needed column

```python
covid_df = pd.DataFrame(df_new["max_infection_rates"])
covid_df.head(10)
```

###Visualizing our final result

###Plotting GDP vs maximum Infection rate

```python
x = df1["GDP per capita"]
y = df1["max_infection_rates"]
sns.scatterplot(x,y)
```

```python
#In above plot We can see the values need different scaling .So, we Will apply log scaling to y
x = df1["GDP per capita"]
y = df1["max_infection_rates"]
sns.scatterplot(x,np.log(y))
plt.show()
```

```python
#RegPlot
x = df1["GDP per capita"]
y = df1["max_infection_rates"]
sns.regplot(x,np.log(y))
plt.show()
#Line fitted, +ve slope seen
```
